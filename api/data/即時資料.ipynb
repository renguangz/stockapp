```python
import requests
import pandas as pd 
import twstock
from twstock import codes
import sqlite3
import json
from ast import literal_eval
```

```python
def get_move(stockid):
    import requests
    import pandas as pd 
    
    url = 'https://tw.stock.yahoo.com/_td-stock/api/resource/FinanceChartService.ApacLibraCharts;autoRefresh=1632713649253;symbols=%5B%22' + stockid + '.TW%22%5D;type=tick?bkt=%5B%22tw-qsp-control-no2-1%22%2C%22tw-qsp-rampup-exp%22%5D&device=desktop&ecma=modern&feature=ecmaModern%2CuseVersionSwitch%2CuseNewQuoteTabColor&intl=tw&lang=zh-Hant-TW&partner=none&prid=9dt91s1gl2eq8&region=TW&site=finance&tz=Asia%2FTaipei&ver=1.2.1149&returnMeta=true'
    res = requests.get(url)

    jd = res.json()['data']
    close = jd[0]['chart']['indicators']['quote'][0]['close']
    timestamp = jd[0]['chart']['timestamp']

    df = pd.DataFrame({'timestamp': timestamp, 'close': close})
    df['time'] = pd.to_datetime(df['timestamp'] + 3600 * 8, unit = 's')
    return df
```

```python
get_move('1101')
```

```python
'''
twstock 最佳五檔
'''

def get_bid(stockid):
    import pandas as pd 
    import twstock
    import json
    
    twstock.realtime.mock = False
    fav_list = ['1101', '2330', '2317', '2498']
    data = twstock.realtime.get(fav_list)
    table = data[stockid]['realtime']
    df = pd.DataFrame(list(table.items()))
    df.columns = ['project', stockid]
    df = df.set_index('project')
    df = df.drop(['latest_trade_price', 'trade_volume', 'accumulate_trade_volume', 'open', 'high', 'low'])
    results = df.to_json()
    parsed = json.loads(results)
    
    return parsed
```

```python
get_bid('2498')
```
