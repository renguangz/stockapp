## 上下在處理重複資料（每天下載完會有很多重複資料）

```python
def crawl_price(stockid, date):
    import requests
    import pandas as pd
    import os
    import time
    import sqlite3
    from talib import abstract
    
    url = 'https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=html&date=' + str(date) + '&stockNo=' + str(stockid)
    data = pd.read_html(requests.get(url).text)[0]
    data.columns = data.columns.droplevel(0)
    data = data.rename(columns={'日期': 'date', '成交股數': 'volume', '開盤價': 'open', '最高價': 'high', '最低價': 'low', '收盤價': 'close'})
    data['volume'] = data['volume'] / 1000
    path = os.path.join('./price_data', 'price_' + str(stockid) + '.csv') # 路徑要換成下面的
    # 還沒有建立個股的csv時，使用下面的
#     data.to_csv(path, index=True)
    conn = sqlite3.connect('price.db')
    sql_table = pd.read_sql('select * from price_' + str(stockid), conn)
    sql_table = sql_table.append(data)
    sql_table = sql_table.set_index('date')
    sql_table = sql_table[sql_table.index.duplicated() == False]
    ma5 = sql_table['close'].rolling(5).mean()
    ma10 = sql_table['close'].rolling(10).mean()
    ma20 = sql_table['close'].rolling(20).mean()
    volumeMa5 = sql_table['volume'].rolling(5).mean()
    volumeMa10 = sql_table['volume'].rolling(10).mean()
    volumeMa20 = sql_table['volume'].rolling(20).mean()
    kd = abstract.STOCH(sql_table)
    slowj = kd['slowk'] * 3 - kd['slowd'] * 2
    sql_table = sql_table.assign(ma5=ma5, ma10=ma10, ma20=ma20, volumeMa5=volumeMa5, volumeMa10=volumeMa10, volumeMa20=volumeMa20, slowk=kd['slowk'], slowd=kd['slowd'], slowj=slowj)
    sql_table.to_sql('price_' + str(stockid), conn, if_exists='replace')
    
#     table = pd.read_csv(path)
#     new_table = table.append(data)
#     new_table = new_table.set_index('date')
#     new_table = new_table[new_table.index.duplicated() == False]
#     ma5 = new_table['close'].rolling(5).mean()
#     ma10 = new_table['close'].rolling(10).mean()
#     ma20 = new_table['close'].rolling(20).mean()
#     new_table = new_table.assign(ma5=ma5, ma10=ma10, ma20=ma20)
#     new_table.to_csv(path, index=True)
#     time.sleep(2)
    return sql_table
    # 先把每個股票資料都存成個別csv
    # 在進行read_csv，爬下一個月份資料，把資料放進read_csv的table裡面，再存回csv
```

```python
download_table = [1101, 2317, 2330, 2498]
for d in download_table:
    table = crawl_price(d, '20210901')
table
```
